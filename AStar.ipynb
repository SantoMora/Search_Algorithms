```python
#Funcion que hace una lista de nodos
def Vertices(L): 
    V = []
    start = []
    goal = []
    for i in range(len(L)):
        for j in range(len(L[i])):
            Ver = []
            Ver.append(i)
            Ver.append(j)
            if L[i][j] != '%': #Pared
                V.append(Ver)
                if L[i][j] == 'P': #Posicion del personaje
                    start = Ver[:]
                elif L[i][j] == '.': #Comida
                    goal = Ver[:]
    return V,start,goal
```

```python
#Busca un nodo en la lista de vertices segun las coordenadas (x,y)
def Buscar(x,y,V): 
    for i in V:
        if x == i[0] and y == i[1]:
            return i
    return False
```

```python
#Funcion que genera el grafo
def Grafo(V,L):
    G = []                       #Si revisamos el mapa buscando las casillas de la derecha 
    A = []                       #y la de abajo podremos encontrar todos los posibles vecinos del grafo
    N2 = []                       
    for i in V:
        x,y = i[0],i[1]
        N2 = Buscar(x,y+1,V)     #Buscamos si hay un nodo abajo
        if N2 != False:
            A.append(Hijo(i,None,0,0))
            A.append(Hijo(N2,None,0,0))
            #A.append(1)
            G.append(A)
        A = []
        N2 = Buscar(x+1,y,V)     #Buscamos si hay un nodo a la derecha
        if N2 != False:
            A.append(Hijo(i,None,0,0))
            A.append(Hijo(N2,None,0,0))
            #A.append(1)
            G.append(A)
        A = []
    return G
```

```python
class Hijo:
    def __init__(self,Nodo,father,g,h,f=None):
        self.Nodo = Nodo
        self.father = father
        self.h = h
        self.g = g
        self.f = g+h
        
    def __repr__(self):
        return str(self.Nodo[0])+" "+str(self.Nodo[1])
    
    def __lt__(self,other):
        return self.f < other.f
    
    def hn(Nodo,goal):#Metodo que retorna el valor heuristico con la distancia Manhattan
        return abs(Nodo.Nodo[0] - goal[0]) + abs(Nodo.Nodo[1] - goal[1])
        
    
```

```python
#Verifica si un nodo esta o no en una lista
def Seen(Nodo,Seen):
    if len(Seen) != 0:
        for i in Seen:
            if Nodo.Nodo == i.Nodo:
                return False
    return True
```

```python
#Busca el nodo con menor valor de f(n) y retorna su posicion en la frontera
def ExtraerNodo(Frontera):
    e = Frontera[0]
    for n in Frontera:
        if n < e:
            e = n
    i = Frontera.index(e)
    return i
```

```python
#Busca los vecinos de un nodo en el grafo
def Sex(Nodo,G,Frontera):
    vecinos = []
    for i in G:
        if i[0].Nodo == Nodo.Nodo:
            vecinos.append(Hijo(i[1].Nodo,Nodo,Nodo.g+1,i[1].hn(goal)))
        elif i[1].Nodo == Nodo.Nodo:
            vecinos.append(Hijo(i[0].Nodo,Nodo,Nodo.g+1,i[0].hn(goal)))
    #Verificar si algun vecino ya esta en la frontera para no meterlo
    for i in vecinos:
        if Seen(i,Frontera):
            Frontera.append(i)
    
def AStar(G,start,goal):
    Frontera = []
    seen = []
    path = []
    Frontera.append(start)
    while Frontera:
        Nodo = Frontera.pop(ExtraerNodo(Frontera)) #Extraer el nodo con menor f(n)
        if Nodo.Nodo == goal.Nodo:
            seen.append(Nodo)
            path.append(Nodo)
            break
        else:
            #Checar si el nodo no se ha visitado ya para buscar sus vecinos
            if Seen(Nodo,seen):
                seen.append(Nodo)
                Sex(Nodo,G,Frontera)
    return Path(seen,path)

#Generar lista con el camino recorrido de start a goal
def Path(seen,path):
    n = seen[:]
    e = n[len(n)-1]
    while e.father != None:
        e = e.father
        path.append(e)
        
    #Ordenar
    path.reverse()
    return path
```
